```
df = pd.read_csv('data/images/findings.csv')
```

```
filenames = os.listdir('data/images')
```

```
df = df[df['Image Index'].isin(filenames)]
```

```
len(df)
```

```
df.head()
```

```
from random import shuffle
```

```
patients = list(df['Patient ID'].unique())
shuffle(patients)

train_patients = patients[:int(0.7 * len(patients))]
val_patients = patients[int(0.7 * len(patients)) : int(0.85 * len(patients))]
test_patients = patients[int(0.85 * len(patients)):]
```

```
train_df = df[df['Patient ID'].isin(train_patients)]
print(len(train_df))
train_df['Finding Labels'].value_counts()
```

```
val_df = df[df['Patient ID'].isin(val_patients)]
print(len(val_df))
val_df['Finding Labels'].value_counts()
```

```
test_df = df[df['Patient ID'].isin(test_patients)]
print(len(test_df))
test_df['Finding Labels'].value_counts()
```

```
train_df = train_df[train_df['Finding Labels'] == 'No Finding']
train_df.to_csv('train.csv', index=False)
val_df.to_csv('val.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
