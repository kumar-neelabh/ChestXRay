```
def unzip_data(base_path):
    import os
    import tarfile
    
    data_path = os.path.join(base_path, 'data/')
    #print('base path : ', base_path)
    #print('data path : ', data_path)
    
    if not os.path.isfile(os.path.join(data_path, 'clean_images.tar.gz')):
        raise FileNotFoundError('clean_images.tar.gz not found in data/')
    
    tar = tarfile.open(os.path.join(data_path, 'clean_images.tar.gz'))
    tar.extractall(path=data_path)
    tar.close()
```

```
unzip_data('/content/gdrive/My Drive/ChestXRay')
```
